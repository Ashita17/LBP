```python
import math
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option("max_seq_items", None)
```

```python
class normal:
    def __init__ (self,nx,ny):
        self.nx=nx
        self.ny=ny

# function for appending the values to the lists

def appending_values(theta,area_fraction,slope,intercept):
    if theta<0:
        theta+=math.pi
    #normal at +90
    n1=normal(-(math.sin(theta)),math.cos(theta))
    NX.append(n1.nx)
    NY.append(n1.ny)
    Area_Fraction.append(area_fraction)
    Direction.append(slope)
    Intercept.append(intercept)

    #normal at -90
    n2=normal(math.sin(theta),-(math.cos(theta)))
    NX.append(n2.nx)
    NY.append(n2.ny)
    Area_Fraction.append(1-area_fraction)
    Direction.append(slope)
    Intercept.append(intercept)

# creating lists for all the columns
NX=[]
NY=[]
Direction=[]
Intercept=[]
Area_Fraction=[]

# for lines going from 2 to 3
        
for i in range(0,11):
    for j in range(0,11):
        if(i!=10):
            slope_here=j/(10-i)
            theta_here=math.atan(slope_here)
            intercept_here=-(i*j)/((10-i)*10)
            area_fraction_here=(0.5*(10-i)*(j))/100

            appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        if(i==10): # edge case
            #normal at +90
            NX.append(-1)
            NY.append(0)
            Direction.append("not defined/vertical line coincident with the right edge")
            Intercept.append("not defined")
            Area_Fraction.append(0)

            #normal at -90
            NX.append(1)
            NY.append(0)
            Direction.append("not defined/vertical line coincident with the right edge")
            Intercept.append("not defined")
            Area_Fraction.append(1)
        
        j+=1
    i+=1

# for lines going from 2 to 4

for i in range(0,11):
    for j in range(0,11):
        if(i!=j):
            slope_here=10/(j-i)
            theta_here=math.atan(slope_here)
            intercept_here=-i/(j-i)
            area_fraction_here=(0.5*10*(20-i-j))/100

            appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        if(i==j): # edge case
            #normal at +90
            NX.append(-1)
            NY.append(0)
            a=str(i/10)
            b="not defined/vertical line at a distance of "+a+" from left edge"
            Direction.append(b)
            Intercept.append("not defined")
            Area_Fraction.append(0)

            #normal at -90
            NX.append(1)
            NY.append(0)
            a=str(i/10)
            b="not defined/vertical line at a distance of "+a+" from left edge"
            Direction.append(b)
            Intercept.append("not defined")
            Area_Fraction.append(1)
        
        j+=1
    i+=1

# for lines going from 2 to 1

for i in range(0,11):
    for j in range(0,11):
        if(i!=0):
            slope_here=-j/i
            theta_here=math.atan(slope_here)
            intercept_here=j/10
            area_fraction_here=(1-((0.5*i*j)/100))

            appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        if(i==0): # edge case
            #normal at +90
            NX.append(-1)
            NY.append(0)
            Direction.append("vertical line coincident with the left edge")
            Intercept.append("not defined")
            Area_Fraction.append(1)

            #normal at -90
            NX.append(1)
            NY.append(0)
            Direction.append("vertical line coincident with the left edge")
            Intercept.append("not defined")
            Area_Fraction.append(0)
        
        j+=1
    i+=1

# for lines going from 3 to 4

for i in range(0,11):
    for j in range(0,11):
        if(i!=10):
            slope_here=(10-j)/(i-10)
            theta_here=math.atan(slope_here)
            intercept_here=1-((i*(10-j))/((i-10)*10))
            area_fraction_here=(0.5*(10-i)*(10-j))/100

            appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        if(i==10): # edge case
            #normal at +90
            NX.append(-1)
            NY.append(0)
            Direction.append("vertical line coincident with the right edge")
            Intercept.append("not defined")
            Area_Fraction.append(0)

            #normal at -90
            NX.append(1)
            NY.append(0)
            Direction.append("vertical line coincident with the right edge")
            Intercept.append("not defined")
            Area_Fraction.append(1)
        
        j+=1
    i+=1

# for lines going from 3 to 1

for i in range(0,11):
    for j in range(0,11):
        slope_here=(j-i)/10
        theta_here=math.atan(slope_here)
        intercept_here=i/10
        area_fraction_here=(0.5*(i+j)*10)/100

        appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        j+=1
    i+=1

# for lines going from 4 to 1

for i in range(0,11):
    for j in range(0,11):
        if(i!=0):
            slope_here=(10-j)/i
            theta_here=math.atan(slope_here)
            intercept_here=j/10
            area_fraction_here=(1-((0.5*i*(10-j))/100))

            appending_values(theta_here,area_fraction_here,slope_here,intercept_here)
        
        if(i==0): # edge case
            #normal at +90
            NX.append(-1)
            NY.append(0)
            Direction.append("vertical line coincident with the left edge")
            Intercept.append("not defined")
            Area_Fraction.append(1)

            #normal at -90
            NX.append(1)
            NY.append(0)
            Direction.append("vertical line coincident with the left edge")
            Intercept.append("not defined")
            Area_Fraction.append(0)
        
        j+=1
    i+=1

# storing the data in dictionary
data={
    'NX':NX,
    'NY':NY,
    'Slope/Nature':Direction,
    'Intercept':Intercept,
    'Area Fraction':Area_Fraction
}

# 220 observations for each loop
#converting dictionary to dataframe
df = pd.DataFrame(data)
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

```python
csv_path = 'output.csv'
df_no_duplicates.to_csv(csv_path, index=True)
```
